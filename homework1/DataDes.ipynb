```python
import wikipediaapi as wikiapi
```

```python
wiki = wikiapi.Wikipedia('ru')
page = wiki.page('Коралловые рифы')
```

Корпус из текстовых документов был создан при помощи Wikipedia-API. Из страницы 'Коралловый риф' в Википедии были изьяты ссылки на другие статьи по схожей тематике. Внизу есть перечень того, о чем в последующем были собраны документы. Это, к примеру, 'Австралия', 'Аталнтический океан', 'Асцидии' и т.д

```python
links = page.links
```

```python
links
```

Всего в статье о Коралловых рифах были найдены 206 ссылок (а значит 206 документов)

```python
len(links)
```

Далее для каждой ссылки находилась статья из Википедии, описание к которой добавлялось в созданный список 'doc'.

```python
doc = []
for title in sorted(links.keys()):
    page_1 = wiki.page(title)
    if page_1.exists():
        doc.append((page_1.summary))
```

Посмотрим суммарное количество слов в списке

```python
a = 0
for st in doc:
    a += int(len(st.split()))
print(a)
```

Узнаем среднее количество слов на документ (общее количество слов делим на количество документов)

```python
a/206
```

Вычислим медиану количества слов в документах

```python
from statistics import median
med = []
for st in doc:
    med.append(int(len(st.split())))
median(med)
```
